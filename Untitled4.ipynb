```python
# @title **Weather Monitoring System Setup**
# @markdown Run this cell first to initialize the system

from datetime import datetime, timedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pytz  # Added for timezone support

# Set your desired timezone (e.g., 'Asia/Kolkata' for India)
TIMEZONE = pytz.timezone('Asia/Kolkata')

def get_local_time():
    """Returns current datetime in your local timezone"""
    return datetime.now(TIMEZONE)

print("‚úÖ Weather monitoring system initialized!")
print("üìÖ Today's date:", get_local_time().strftime("%Y-%m-%d"))
print("üïí Current time:", get_local_time().strftime("%H:%M:%S"))
print("üåê Timezone:", TIMEZONE.zone)
```

```python
# @title **Sensor Configuration**
# @markdown Set your base temperature and humidity parameters

from datetime import datetime
import pytz

# Timezone configuration (same as setup box)
TIMEZONE = pytz.timezone('Asia/Kolkata')  # Change to your timezone

# --- Configurable Parameters ---
base_temp = 29  # @param {type:"slider", min:15, max:40, step:0.5}
base_humidity = 70  # @param {type:"slider", min:30, max:95, step:1}
num_readings = 10  # @param {type:"slider", min:5, max:20, step:1}
simulation_speed = 1  # @param {type:"slider", min:0.1, max:5, step:0.1}

# --- Input Validation ---
if not (15 <= base_temp <= 40):
    raise ValueError("Temperature must be between 15¬∞C and 40¬∞C")
if not (30 <= base_humidity <= 95):
    raise ValueError("Humidity must be between 30% and 95%")
if not (5 <= num_readings <= 20):
    raise ValueError("Number of readings must be between 5 and 20")

# --- Configuration Summary ---
print(f"‚öôÔ∏è Configuration set at {datetime.now(TIMEZONE).strftime('%Y-%m-%d %H:%M:%S')}")
print(f"üå°Ô∏è Base temperature: {base_temp}¬∞C (Range: 15-40¬∞C)")
print(f"üíß Base humidity: {base_humidity}% (Range: 30-95%)")
print(f"üìä Number of readings: {num_readings} (5-20)")
print(f"‚è±Ô∏è Simulation speed: {simulation_speed}x")

# Store configuration in a dictionary for easy access
config = {
    'base_temp': base_temp,
    'base_humidity': base_humidity,
    'num_readings': num_readings,
    'simulation_speed': simulation_speed,
    'timezone': TIMEZONE.zone,
    'config_time': datetime.now(TIMEZONE)
}
```

```python
# @title **DHT11 Sensor Simulator**
# @markdown This cell simulates the sensor behavior with realistic daily patterns

import numpy as np
from datetime import datetime
import pytz

class DHT11Simulator:
    def __init__(self, temp, humidity, timezone='Asia/Kolkata'):
        """
        Initialize the sensor simulator with base values
        Args:
            temp: Base temperature in ¬∞C
            humidity: Base humidity in %
            timezone: Timezone string (e.g., 'Asia/Kolkata')
        """
        self.base_temp = temp
        self.base_humidity = humidity
        self.timezone = pytz.timezone(timezone)
        self.last_read_time = None
        self.reading_count = 0

    def read(self):
        """Simulate a sensor reading with realistic daily patterns and noise"""
        # Get current time in specified timezone
        now = datetime.now(self.timezone)
        hour = now.hour + now.minute/60  # More precise hour fraction

        # Calculate daily patterns (more accurate with hour fraction)
        temp_variation = -3 * np.cos(2 * np.pi * (hour - 14) / 24)  # Colder at night
        hum_variation = 10 * np.sin(2 * np.pi * (hour - 6) / 24)    # More humid in morning

        # Add realistic noise (more variation during daytime)
        temp_noise = np.random.uniform(-0.8 if 6<hour<18 else -0.3,
                                       0.8 if 6<hour<18 else 0.3)
        hum_noise = np.random.uniform(-3 if 6<hour<18 else -1,
                                      3 if 6<hour<18 else 1)

        # Calculate current values
        current_temp = self.base_temp + temp_variation + temp_noise
        current_hum = self.base_humidity + hum_variation + hum_noise

        # Ensure realistic physical limits
        current_temp = np.clip(current_temp, 10, 40)
        current_hum = np.clip(current_hum, 30, 95)

        # Track reading statistics
        self.last_read_time = now
        self.reading_count += 1

        return current_temp, current_hum

# Initialize sensor with config from previous cell
sensor = DHT11Simulator(
    temp=base_temp,
    humidity=base_humidity,
    timezone=TIMEZONE.zone
)

print(f"üå°Ô∏è DHT11 Sensor Simulator ready at {datetime.now(TIMEZONE).strftime('%H:%M:%S')}")
print(f"   - Base Temp: {base_temp}¬∞C")
print(f"   - Base Humidity: {base_humidity}%")
print(f"   - Timezone: {TIMEZONE.zone}")
```

```python
# @title **Data Collection**
# @markdown Collecting and processing sensor readings with timezone support

import time
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

def classify_weather(temp, hum):
    """Classify weather condition based on temperature and humidity"""
    if temp > 32: return "üî• Very Hot"
    elif temp > 28: return "‚òÄÔ∏è Hot"
    elif temp < 15: return "‚ùÑÔ∏è Cold"
    elif hum > 80: return "üíß Humid"
    else: return "üòä Pleasant"

# Initialize empty DataFrame with proper types
df = pd.DataFrame(columns=['timestamp', 'temperature', 'humidity', 'condition'])
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.tz_localize(TIMEZONE)

# Collect readings with progress tracking
print(f"‚è≥ Collecting {num_readings} readings...")
for i in range(num_readings):
    # Get simulated reading
    temp, hum = sensor.read()

    # Calculate timestamp with proper timezone (fixed parenthesis here)
    reading_time = datetime.now(TIMEZONE) - timedelta(minutes=(num_readings-i)*5)

    # Add to DataFrame
    df.loc[len(df)] = {
        'timestamp': reading_time,
        'temperature': temp,
        'humidity': hum,
        'condition': classify_weather(temp, hum)
    }

    # Visual progress indicator
    if (i+1) % max(1, num_readings//10) == 0 or (i+1) == num_readings:
        print(f"   {i+1}/{num_readings} readings collected", end='\r')

    time.sleep(0.1 * config['simulation_speed'])  # Adjustable delay

# Data processing
df = df.sort_values('timestamp').reset_index(drop=True)
df['temp_change'] = df['temperature'].diff()
df['hum_change'] = df['humidity'].diff()

print(f"\n‚úÖ Successfully collected {len(df)} valid readings")
print(f"   Time range: {df['timestamp'].iloc[0].strftime('%H:%M')} to {df['timestamp'].iloc[-1].strftime('%H:%M')}")
print(f"   Avg Temp: {df['temperature'].mean():.1f}¬∞C | Avg Hum: {df['humidity'].mean():.1f}%")

# Display results
def color_conditions(val):
    colors = {
        "üî• Very Hot": "#FF6B6B",
        "‚òÄÔ∏è Hot": "#FFA500",
        "üòä Pleasant": "#A8E6CF",
        "üíß Humid": "#64B5F6",
        "‚ùÑÔ∏è Cold": "#BBE6FF"
    }
    return f'background-color: {colors.get(val, "#FFFFFF")}'

styled_df = df.tail(3).style\
    .format({
        'temperature': '{:.1f}¬∞C',
        'humidity': '{:.1f}%',
        'temp_change': '{:+.1f}¬∞C',
        'hum_change': '{:+.1f}%'
    })\
    .applymap(color_conditions, subset=['condition'])\
    .set_caption(f"Latest Readings ({TIMEZONE.zone})")

display(styled_df)
```

```python
# @title **Weather Forecast Prediction**
# @markdown Generating timezone-aware 5-hour predictions

def predict_weather(last_temp, last_hum, n_hours=5):
    """Generate weather predictions with timezone awareness"""
    predictions = []
    now = datetime.now(TIMEZONE)

    for hour in range(1, n_hours+1):
        future_time = now + timedelta(hours=hour)
        future_hour = future_time.hour + future_time.minute/60  # More precise

        # Temperature prediction with daily pattern
        temp_change = (-0.5 if 0 <= future_hour < 6 else 0.3) * hour
        temp_noise = np.random.uniform(-0.8 if 6<future_hour<18 else -0.3,
                                      0.8 if 6<future_hour<18 else 0.3)
        pred_temp = last_temp + temp_change + temp_noise

        # Humidity prediction with daily pattern
        hum_change = (2 if 4 <= future_hour < 12 else -1) * hour
        hum_noise = np.random.uniform(-3 if 6<future_hour<18 else -1,
                                     3 if 6<future_hour<18 else 1)
        pred_hum = last_hum + hum_change + hum_noise

        # Apply realistic limits
        pred_temp = np.clip(pred_temp, 10, 40)
        pred_hum = np.clip(pred_hum, 30, 95)

        predictions.append({
            'time': future_time,
            'display_time': future_time.strftime("%-I %p"),  # Hour without leading zero
            'temperature': pred_temp,
            'humidity': pred_hum,
            'condition': classify_weather(pred_temp, pred_hum)
        })

    return pd.DataFrame(predictions)

# Generate forecast using latest readings
forecast = predict_weather(
    last_temp=df['temperature'].iloc[-1],
    last_hum=df['humidity'].iloc[-1]
)

# Enhanced display
def color_by_temp(temp):
    """Color gradient based on temperature"""
    if temp > 32: return 'background-color: #ff6b6b; color: white'  # Very hot
    elif temp > 28: return 'background-color: #ffa07a'  # Hot
    elif temp < 15: return 'background-color: #add8e6'  # Cold
    else: return 'background-color: #98fb98'  # Pleasant

print(f"üîÆ 5-Hour Forecast for {TIMEZONE.zone}")
display(
    forecast.style
    .format({
        'temperature': '{:.1f}¬∞C',
        'humidity': '{:.1f}%'
    })
    .applymap(color_by_temp, subset=['temperature'])
    .set_caption(f"Weather Forecast starting {datetime.now(TIMEZONE).strftime('%b %d, %H:%M')}")
    .set_properties(**{
        'text-align': 'center',
        'font-weight': 'bold'
    })
    .bar(subset=['humidity'], color='#5f9ea0')
)
```

```python
# @title **Data Visualization**
# @markdown Professional timezone-aware weather visualization

import matplotlib.dates as mdates
from matplotlib import rcParams

# ===== SETUP =====
# Configure global plot settings
rcParams.update({
    'font.family': 'DejaVu Sans',
    'axes.titlepad': 20,
    'axes.labelpad': 10,
    'axes.titlesize': 16,
    'axes.labelsize': 14,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'legend.fontsize': 12
})

# Ensure all timestamps are timezone-aware and consistent
def ensure_timezone(series, timezone):
    """Convert all timestamps to the specified timezone"""
    if series.dt.tz is None:
        return series.dt.tz_localize(timezone)
    return series.dt.tz_convert(timezone)

# Convert all timestamps to the same timezone
df['timestamp'] = ensure_timezone(df['timestamp'], TIMEZONE)
forecast['time'] = ensure_timezone(forecast['time'], TIMEZONE)

# Create figure with constrained layout
fig = plt.figure(figsize=(16, 12), constrained_layout=True)
gs = fig.add_gridspec(2, 1, hspace=0.3)

# ===== TEMPERATURE PLOT =====
ax1 = fig.add_subplot(gs[0])
# Plot actual and forecast data
ax1.plot(df['timestamp'], df['temperature'], 'b-o',
        markersize=10, linewidth=2.5,
        markerfacecolor='white', markeredgewidth=2,
        label='Actual Data')

ax1.plot(forecast['time'], forecast['temperature'], 'r--o',
        markersize=10, linewidth=2.5,
        markerfacecolor='white', markeredgewidth=2,
        label='Forecast')

# Visual bridge between actual and forecast
ax1.fill_between([df['timestamp'].iloc[-1], forecast['time'].iloc[0]],
                [df['temperature'].iloc[-1], forecast['temperature'].iloc[0]],
                color='orange', alpha=0.1)

# Threshold zones and lines
ax1.axhspan(30, 40, facecolor='#ffcccc', alpha=0.1)
ax1.axhspan(10, 15, facecolor='#ccccff', alpha=0.1)
ax1.axhline(30, color='red', linestyle='-.', alpha=0.7, linewidth=1.5, label='Hot Threshold (30¬∞C)')
ax1.axhline(15, color='blue', linestyle='-.', alpha=0.7, linewidth=1.5, label='Cold Threshold (15¬∞C)')

# Formatting
ax1.set_title(f'Temperature Trend - {TIMEZONE.zone}', pad=25)
ax1.set_ylabel('Temperature (¬∞C)')
ax1.grid(True, alpha=0.2, linestyle='--')

# ===== HUMIDITY PLOT =====
ax2 = fig.add_subplot(gs[1], sharex=ax1)
ax2.plot(df['timestamp'], df['humidity'], 'g-o',
        markersize=10, linewidth=2.5,
        markerfacecolor='white', markeredgewidth=2,
        label='Actual Data')

ax2.plot(forecast['time'], forecast['humidity'], 'm--o',
        markersize=10, linewidth=2.5,
        markerfacecolor='white', markeredgewidth=2,
        label='Forecast')

# Threshold zones and lines
ax2.axhspan(80, 95, facecolor='#ccffff', alpha=0.1)
ax2.axhspan(30, 40, facecolor='#ffebcc', alpha=0.1)
ax2.axhline(80, color='teal', linestyle='-.', alpha=0.7, linewidth=1.5, label='Humid Threshold (80%)')
ax2.axhline(30, color='peru', linestyle='-.', alpha=0.7, linewidth=1.5, label='Dry Threshold (30%)')

# Formatting
ax2.set_title(f'Humidity Trend - {TIMEZONE.zone}', pad=25)
ax2.set_ylabel('Humidity (%)')
ax2.set_xlabel('Time')
ax2.grid(True, alpha=0.2, linestyle='--')

# ===== COMMON FORMATTING =====
for ax in [ax1, ax2]:
    # X-axis formatting
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%a %H:%M', tz=TIMEZONE))
    ax.xaxis.set_major_locator(mdates.HourLocator(interval=3))

    # Rotate and align ticks
    plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

    # Legend
    ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left',
             facecolor='white', framealpha=0.8)

# ===== FINAL TOUCHES =====
fig.suptitle(f'Weather Station Analysis - {datetime.now(TIMEZONE).strftime("%Y-%m-%d %H:%M")}',
            y=1.02, fontsize=14, color='#555555')

plt.show()
```

```python
# @title **Export Weather Data**
# @markdown Save your collected data with proper timezone handling

from google.colab import files
from datetime import datetime

def prepare_export_data(actual_df, forecast_df):
    """Prepare data for export with consistent timezone handling"""
    # Create copies to avoid modifying original dataframes
    actual_export = actual_df.copy()
    forecast_export = forecast_df.copy()

    # Standardize column names and add type indicators
    actual_export['type'] = 'Actual'
    forecast_export['type'] = 'Forecast'

    # Rename columns for consistency
    forecast_export = forecast_export.rename(columns={'time': 'timestamp'})

    # Select common columns
    common_cols = ['timestamp', 'temperature', 'humidity', 'condition', 'type']

    # Combine data with proper timezone handling
    combined = pd.concat([
        actual_export[common_cols],
        forecast_export[common_cols]
    ])

    # Ensure timestamp is timezone-aware and formatted consistently
    if combined['timestamp'].dt.tz is None:
        combined['timestamp'] = combined['timestamp'].dt.tz_localize(TIMEZONE)
    else:
        combined['timestamp'] = combined['timestamp'].dt.tz_convert(TIMEZONE)

    # Add human-readable date/time columns
    combined['date'] = combined['timestamp'].dt.strftime('%Y-%m-%d')
    combined['time'] = combined['timestamp'].dt.strftime('%H:%M:%S')
    combined['timezone'] = TIMEZONE.zone

    return combined.sort_values('timestamp')

# Prepare data for export
export_data = prepare_export_data(df, forecast)

# Save options
format = 'CSV'  # @param ['CSV', 'Excel', 'JSON']
filename = f"weather_data_{datetime.now(TIMEZONE).strftime('%Y%m%d_%H%M')}"

# Export data
try:
    if format == 'CSV':
        export_data.to_csv(f'{filename}.csv', index=False)
        files.download(f'{filename}.csv')
    elif format == 'Excel':
        export_data.to_excel(f'{filename}.xlsx', index=False)
        files.download(f'{filename}.xlsx')
    elif format == 'JSON':
        export_data.to_json(f'{filename}.json', orient='records', date_format='iso')
        files.download(f'{filename}.json')

    print(f"‚úÖ Data exported successfully as {format}")
    print(f"üìÅ Filename: {filename}")
    print(f"üìä Records: {len(export_data)} total ({len(df)} actual, {len(forecast)} forecast)")
    print(f"‚è± Time range: {export_data['timestamp'].min().strftime('%Y-%m-%d %H:%M')} to "
          f"{export_data['timestamp'].max().strftime('%Y-%m-%d %H:%M')} {TIMEZONE.zone}")

except Exception as e:
    print(f"‚ùå Export failed: {str(e)}")
    print("‚ÑπÔ∏è Please check if the file is open in another program and try again")
```
