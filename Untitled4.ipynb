```python
# @title **Weather Monitoring System Setup**
# @markdown Run this cell first to initialize the system

from datetime import datetime, timedelta
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("‚úÖ Weather monitoring system initialized!")
print("üìÖ Today's date:", datetime.now().strftime("%Y-%m-%d"))
```

```python
# @title **Sensor Configuration**
# @markdown Set your base temperature and humidity

base_temp = 29  # @param {type:"slider", min:15, max:40, step:0.5}
base_humidity = 70  # @param {type:"slider", min:30, max:95, step:1}
num_readings = 10  # @param {type:"slider", min:5, max:20, step:1}

print(f"‚öôÔ∏è Configuration set to:")
print(f"- Base temperature: {base_temp}¬∞C")
print(f"- Base humidity: {base_humidity}%")
print(f"- Number of readings: {num_readings}")
```

```python
# @title **DHT11 Sensor Simulator**
# @markdown This cell simulates the sensor behavior

class DHT11Simulator:
    def __init__(self, temp, humidity):
        self.base_temp = temp
        self.base_humidity = humidity

    def read(self):
        hour = datetime.now().hour
        # Daily temperature pattern (colder at night)
        temp_variation = -3 * np.cos(2 * np.pi * (hour - 14) / 24)
        # Daily humidity pattern (more humid in morning)
        hum_variation = 10 * np.sin(2 * np.pi * (hour - 6) / 24)

        current_temp = self.base_temp + temp_variation + np.random.uniform(-0.5, 0.5)
        current_hum = self.base_humidity + hum_variation + np.random.uniform(-2, 2)

        # Ensure realistic values
        current_temp = max(10, min(40, current_temp))
        current_hum = max(30, min(95, current_hum))

        return current_temp, current_hum

sensor = DHT11Simulator(base_temp, base_humidity)
print("üå°Ô∏è DHT11 Sensor Simulator ready!")
```

```python
# @title **Data Collection**
# @markdown Collecting sensor readings...

import time  # Added this import
from datetime import datetime, timedelta
import numpy as np
import pandas as pd

readings = []
for i in range(num_readings):
    temp, hum = sensor.read()
    readings.append({
        'timestamp': datetime.now() - timedelta(minutes=(num_readings-i)*5),
        'temperature': temp,
        'humidity': hum,
        'condition': ''
    })
    time.sleep(0.1)  # Simulate reading delay

df = pd.DataFrame(readings)

def classify_weather(temp, hum):
    if temp > 32: return "üî• Very Hot"
    elif temp > 28: return "‚òÄÔ∏è Hot"
    elif temp < 15: return "‚ùÑÔ∏è Cold"
    elif hum > 80: return "üíß Humid"
    else: return "üòä Pleasant"

df['condition'] = df.apply(lambda x: classify_weather(x['temperature'], x['humidity']), axis=1)

print(f"üìä Collected {len(df)} readings:")
display(df.tail(3).style.set_caption("Latest Readings"))
```

```python
# @title **Weather Forecast Prediction**
# @markdown Generating 5-hour predictions

def predict_weather(last_temp, last_hum, n_hours=5):
    predictions = []
    for hour in range(1, n_hours+1):
        future_time = datetime.now() + timedelta(hours=hour)
        future_hour = future_time.hour

        # Temperature prediction (cools at night, warms during day)
        temp_change = (-0.5 if 0 <= future_hour < 6 else 0.3) * hour
        pred_temp = last_temp + temp_change + np.random.uniform(-0.5, 0.5)

        # Humidity prediction (increases in morning)
        hum_change = (2 if 4 <= future_hour < 12 else -1) * hour
        pred_hum = last_hum + hum_change + np.random.uniform(-2, 2)

        predictions.append({
            'time': future_time.strftime("%I %p").lstrip("0"),
            'temperature': max(10, min(40, pred_temp)),
            'humidity': max(30, min(95, pred_hum)),
            'condition': classify_weather(pred_temp, pred_hum)
        })
    return pd.DataFrame(predictions)

forecast = predict_weather(df['temperature'].iloc[-1], df['humidity'].iloc[-1])

print("üîÆ 5-Hour Weather Forecast:")
display(forecast.style
       .format({'temperature': '{:.1f}¬∞C', 'humidity': '{:.1f}%'})
       .set_caption("Upcoming Conditions")
       .background_gradient(cmap='Blues'))
```

```python
# @title **Data Visualization**
# @markdown Interactive weather trends

plt.figure(figsize=(12, 8))

# Convert forecast times to datetime objects
forecast_times = [datetime.now() + timedelta(hours=h) for h in range(1,6)]

# Temperature plot
plt.subplot(2, 1, 1)
plt.plot(df['timestamp'], df['temperature'], 'b-o', markersize=8, label='Actual')
plt.plot(forecast_times, forecast['temperature'], 'r--o', markersize=8, label='Forecast')
plt.title('Temperature Trend', fontsize=14, pad=20)
plt.ylabel('¬∞C', fontsize=12)
plt.xticks(rotation=45)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)

# Humidity plot
plt.subplot(2, 1, 2)
plt.plot(df['timestamp'], df['humidity'], 'g-o', markersize=8, label='Actual')
plt.plot(forecast_times, forecast['humidity'], 'm--o', markersize=8, label='Forecast')
plt.title('Humidity Trend', fontsize=14, pad=20)
plt.ylabel('%', fontsize=12)
plt.xticks(rotation=45)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# @title **Weather Forecast Prediction**
# @markdown Generating 5-hour predictions

def predict_weather(last_temp, last_hum, n_hours=5):
    predictions = []
    for hour in range(1, n_hours+1):
        future_time = datetime.now() + timedelta(hours=hour)

        # Temperature prediction
        temp_change = (-0.5 if 0 <= future_time.hour < 6 else 0.3) * hour
        pred_temp = last_temp + temp_change + np.random.uniform(-0.5, 0.5)

        # Humidity prediction
        hum_change = (2 if 4 <= future_time.hour < 12 else -1) * hour
        pred_hum = last_hum + hum_change + np.random.uniform(-2, 2)

        predictions.append({
            'time': future_time,  # Store as datetime object
            'temperature': max(10, min(40, pred_temp)),
            'humidity': max(30, min(95, pred_hum)),
            'condition': classify_weather(pred_temp, pred_hum)
        })
    return pd.DataFrame(predictions)

forecast = predict_weather(df['temperature'].iloc[-1], df['humidity'].iloc[-1])

print("üîÆ 5-Hour Weather Forecast:")
display(forecast.assign(
    display_time=forecast['time'].dt.strftime("%I %p").str.lstrip("0")
).style
       .format({'temperature': '{:.1f}¬∞C', 'humidity': '{:.1f}%'})
       .set_caption("Upcoming Conditions")
       .background_gradient(cmap='Blues'))
```
